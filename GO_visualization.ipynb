```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DB00927_df = pd.read_csv('DB00927_top10.csv')
DB00927_df.head()
```

```python
DB00927_df_sorted= DB00927_df.sort_values('pvalue',ascending=False)
plt.barh(DB00927_df_sorted.Description, DB00927_df_sorted.pvalue)
plt.xlabel("p-value")
plt.ylabel("Biological Processes")
plt.title("Pathways enriched for gene variants associated with breast cancer in Famotidine")
plt.show()
```
